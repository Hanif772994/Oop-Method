```python
# encapsulation

class bank:
    def init_amount(self, _balance):
        self._balance= _balance
        return _balance
    
    def deposit(self, amount):
        self._balance += amount
        return self._balance
    def withdraw(self, money):
        if self._balance < money:
            print("you are gadha")
        else:
            self._balance -= money
            return self._balance

        
b= bank()
print(b.init_amount(1000000))
print(b.deposit(10))
print(b.withdraw(1000))
```
